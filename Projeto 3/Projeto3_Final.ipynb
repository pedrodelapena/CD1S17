## Estudo da Distribuição da Média Amostral  $\bar{X}$  via SIMULAÇÃO

### Distribuição escolhida - Poisson, com E(X) = Var(X) = 7

#### 1) Construindo a população

```python
poiss = stats.poisson.rvs(mu=7, loc=0, size=1333337)
```

```python
pm = poiss.mean()
pv = poiss.var()
faixa = range(0,26,1)
plt.hist(poiss, bins=faixa, normed=True, color="indigo", alpha=0.6)
plt.grid()
plt.xlabel = ("Valor da variável")
plt.ylabel = ("Frequência absoluta")
print("A média da distribuição teórica é:", pm)
print("")
print("A variância da distribuição teórica é:", pv)
```

#### 2)

#### n = 2

```python
medias_n2 = []
for i in range(10000):
    rc = np.random.choice(poiss, size=2)
    medias_n2.append(rc.mean())
    i+= 1

media_das_medias_n2 = np.mean(medias_n2)

var_das_medias_n2 = np.var(medias_n2)

print("Média das médias com n = 2:",media_das_medias_n2)
print("Variância das médias com n = 2:",var_das_medias_n2)
```

```python
plt.hist(medias_n2, color = 'red', alpha =0.7, normed = True, bins = 20)
plt.grid()
plt.title("n = 2")
```

## Estudo da Distribuição de  $\frac{\bar{X}-\mu}{\sigma\hspace{1mm}/\sqrt{n}}$  e de $\frac{\bar{X}-\mu}{S\hspace{1mm}/\sqrt{n}}$ via SIMULAÇÃO

```python
norm = stats.norm.rvs(scale=2, loc=10, size=1333337)
```

```python
nm = norm.mean()
nv = norm.var()
m = 10
sig = 2
plt.hist(norm, bins=20, normed=True, color="indigo", alpha=0.6)
plt.grid()
plt.xlabel = ("Valor da variável")
plt.ylabel = ("Frequência absoluta")
print("A média da distribuição teórica é:", nm)
print("")
print("A variância da distribuição teórica é:", nv)
```

### Funções para o cálculo das médias e desvios padrões

```python
def med_amostral(n):
    mednorm=[]
    for i in range (n):
        mednorm.append(np.random.choice(norm))
    media = (sum(mednorm))/n
    std = np.std(mednorm)
    return media, std

#n=4
medias_n4 = []
for i in range (10000):
    medias_n4.append(med_amostral(4)[0])

std_n4 = []
for i in range (10000):
    std_n4.append(med_amostral(4)[1])
    
#n=20
medias_n20 = []
for i in range (10000):
    medias_n20.append(med_amostral(20)[0])

std_n20 = []
for i in range (10000):
    std_n20.append(med_amostral(20)[1])

#n=50
medias_n50 = []
for i in range (10000):
    medias_n50.append(med_amostral(50)[0])

std_n50 = []
for i in range (10000):
    std_n50.append(med_amostral(50)[1])
    
#n=300
medias_n300 = []
for i in range (10000):
    medias_n300.append(med_amostral(300)[0])

std_n300 = []
for i in range (10000):
    std_n300.append(med_amostral(300)[1])
    
```

```python
def dist_z (nm, m, sig, n):
    z = (nm-m)/(sig/((n)**0.5))
    return z

def dist_t (nm, m, s, n):
    t = (nm-m)/(s/((n)**0.5))
    return t
```

### Distribuição da média - n = 4

```python
import math

z = []
t = []
x = np.linspace(-4, 4, 10000)

for i in range (10000):
    z.append(dist_z(medias_n4[i], m, sig, 4))

for i in range (10000):
    t.append(dist_t(medias_n4[i], m, std_n4[i], 4))


fig = plt.figure(figsize=(16,5))
mu = 0
stdd = 1
x = np.linspace(-8, 8, 10000)

plt.subplot(121)
plt.hist(z,bins=25,normed=True, color = "indigo", alpha = 0.6, label="aaaa")
plt.plot(x,stats.norm.pdf(x, mu, stdd), lw=3, color='black')
plt.title("Histograma - Z")
plt.ylim(0, 0.45)
plt.grid()

plt.subplot(122)
plt.hist(t, bins=200, normed=True, color = "lime", alpha = 0.6)
plt.plot(x, stats.t.pdf(x,df=3), lw=3, color='black')
plt.title("Histograma - T")
plt.xlim(-15,15)
plt.ylim(0,0.45)
plt.grid()

plt.tight_layout()
plt.show()
```

## Conclusão

As distribuições Z, que utilizam do sigma da população real, permanecem com o formato constante do histograma e se assemelham muito com a distribuição normal, sendo independentes do aumento ou redução da quantidade de amostras (n). Já as distribuições T utilizam o desvio padrão da população observada e portanto, o número de amostras faz com que a distribuiçao seja alterada. Para mais amostras observadas, a variância tende a diminuir e os valores passam a ser mais frequentes próximos à media da distribuição.
